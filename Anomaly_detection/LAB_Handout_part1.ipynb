#  Anomaly Detection

**Anomaly detection** (also called **outlier detection**) identifies data points that deviate significantly from the majority of the dataset.

Anomalies can signal **errors, fraud, system failures, or rare events**.

In this method, an **anomaly** is a point that is **far away from its neighbors**.

We can measure this distance using **Euclidean distance** and detect outliers based on **k-nearest neighbors (KNN)** distances.

---

### ğŸ§© Steps

1. Choose a value of **k** (number of neighbors).  
2. For each data point, calculate the **distance to its k nearest neighbors**.  
3. Points with **large average distances** are likely **anomalies**.  
4. Visualize or rank points by distance.

### ğŸ“ Notes

- Works well for **non-parametric** data (no assumption of distribution).
